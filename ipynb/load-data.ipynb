## Setup

```python
# Import libraries needed

import pandas as pd
import requests
import io
```

```python
github_repo = "https://raw.githubusercontent.com/spencermarley/smu-cs605-nlp-semeval/main/data/"
```

## Import Data

### Laptops (2014)

```python
# 2014 Laptops Trial
df_2014_laptops_trial = pd.read_csv(io.StringIO(requests.get(github_repo+"2014/laptops/trial"+".csv").content.decode('utf-8')), skiprows=2, names=['id','number','from','polarity','term','to','text'])
# df_2014_laptops_trial.head()

# 2014 Laptops Train
df_2014_laptops_train = pd.read_csv(io.StringIO(requests.get(github_repo+"2014/laptops/train"+".csv").content.decode('utf-8')), skiprows=2, names=['id','number','from','polarity','term','to','text'])
# df_2014_laptops_train.head()

# 2014 Laptops Train v2
df_2014_laptops_train_v2 = pd.read_csv(io.StringIO(requests.get(github_repo+"2014/laptops/train-v2"+".csv").content.decode('utf-8')), skiprows=2, names=['id','number','from','polarity','term','to','text'])
# df_2014_laptops_train_v2.head()

# 2014 Laptops Test A
df_2014_laptops_test_a = pd.read_csv(io.StringIO(requests.get(github_repo+"2014/laptops/test-a"+".csv").content.decode('utf-8')), skiprows=2, names=['id','text'])
# df_2014_laptops_test_a.head()

# 2014 Laptops Test B
df_2014_laptops_test_b = pd.read_csv(io.StringIO(requests.get(github_repo+"2014/laptops/test-b"+".csv").content.decode('utf-8')), skiprows=2, names=['id','number','from','term','to','text'])
# df_2014_laptops_test_b.head()
```
